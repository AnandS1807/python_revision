Data Types
Python has various built-in data types that allow us to store different kinds of values. Hereâ€™s a brief overview of the main data types:

int (Integer): Represents whole numbers, both positive and negative, without any decimal points. Example: x = 5, y = -42
float (Floating Point Number): Represents real numbers (numbers with a decimal point). Example: x = 3.14, y = -0.001

str (String): Represents a sequence of characters enclosed in single, double, or triple quotes. Example: name = "Alice", greeting = 'Hello World'
bool (Boolean): Represents one of two values: True or False. Often used in conditional statements. Example: is_active = True, is_valid = False

```python
def add_item(item, box=None):
    if box is None : box = []
    box.append(item)
    return box

print(add_item('a'))
print(add_item('b'))
```

```python
#converting python objects into json strings using dumps method 

import json 
python_data = {'name': 'ROnin', 'age':21}
json_data = json.dumps(python_data)
print(python_data)
print(json_data)

#printing the json striung using indent - i.e called pretty-printing
json_data2 = json.dumps(python_data, indent =4 )
print(json_data2)


#printing the json string with indent and sorted in alhbetical order 
json_data3 = json.dumps(python_data, indent=4, sort_keys = True)
print(json_data3)
```
