## 2️⃣ Shallow Copy

A shallow copy creates a new outer object, but inner objects are shared.

```python
import copy

b = copy.copy(a)   # generic
b = a.copy()       # lists, dicts, sets
b = a[:]           # lists
```

```python
#exmaple 
a = [[1, 2], [3, 4]]
b = a.copy()

b[0].append(99)

print(a)  # [[1, 2, 99], [3, 4]]
print(b)  # [[1, 2, 99], [3, 4]]
```
